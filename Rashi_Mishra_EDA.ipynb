```python
customer_data = merged_df.groupby('CustomerID').agg(Recency=('TransactionDate', lambda x: (merged_df['TransactionDate'].max() - x.max()).days),
                                                    Frequency=('TransactionID', 'count'),
                                                    Monetary=('TotalValue', 'sum'))

customer_data['CustomerSegment'] = pd.qcut(customer_data['Monetary'], 4, labels=['Low', 'Medium', 'High', 'VIP'])
customer_data
```

```python
# Example logic for R, F, and M scores
customer_data['R_Score'] = pd.qcut(customer_data['Recency'], q=5, labels=range(5, 0, -1)).astype(int)
customer_data['F_Score'] = pd.qcut(customer_data['Frequency'], q=5, labels=range(1, 6)).astype(int)
customer_data['M_Score'] = pd.qcut(customer_data['Monetary'], q=5, labels=range(1, 6)).astype(int)

# Group by R and F scores and calculate mean of M scores
rfm_heatmap = customer_data.groupby(['R_Score', 'F_Score']).agg({'M_Score': 'mean'}).reset_index()

# Pivot to create a grid
rfm_pivot = rfm_heatmap.pivot(index='R_Score', columns='F_Score', values='M_Score')

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    rfm_pivot,
    annot=True,
    fmt='.1f',
    cmap='YlGnBu',
    cbar_kws={'label': 'Monetary Score'}
)
plt.title('RFM Grid Heatmap (Mean Monetary Scores)', fontsize=14)
plt.xlabel('Frequency Score (F)', fontsize=12)
plt.ylabel('Recency Score (R)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
